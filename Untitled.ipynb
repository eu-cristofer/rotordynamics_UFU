```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# System parameters for a 4-DOF system
m = np.array([1.0, 1.5, 2.0, 2.5])  # masses (kg)
k = np.array([20.0, 25.0, 30.0, 35.0])  # stiffness (N/m)
c = np.array([1.0, 1.2, 1.5, 1.8])  # damping coefficients (Ns/m)
F = np.array([1.0, 0.5, 0.2, 0.1])  # force amplitudes (N)
omega = 2.0  # force frequency (rad/s)

# Mass matrix (diagonal for simplicity)
M = np.diag(m)

# Damping matrix (diagonal for simplicity)
C = np.diag(c)

# Stiffness matrix (diagonal for simplicity)
K = np.diag(k)

# State-space representation
A = np.block([
    [np.zeros((4, 4)), np.eye(4)],
    [-np.linalg.inv(M) @ K, -np.linalg.inv(M) @ C]
])

B = np.block([
    [np.zeros((4, 4))],
    [np.linalg.inv(M)]
])

# Initial conditions
z0 = np.zeros(8)  # [initial displacements, initial velocities]

# Time vector
t = np.linspace(0, 10, 100)

# Define the forcing function
def forcing_function(t, omega):
    return np.sin(omega * t)

# Define the state-space model including the sinusoidal force
def state_space_model(z, t, A, B, F, omega):
    u = F * forcing_function(t, omega)
    dzdt = A @ z + B @ u
    return dzdt

# Solve the differential equation
solution = odeint(state_space_model, z0, t, args=(A, B, F, omega))

# Extract displacements and velocities
x = solution[:, :4]
v = solution[:, 4:]

# Plot the results
plt.figure()
for i in range(4):
    plt.subplot(4, 1, i+1)
    plt.plot(t, x[:, i], label=f'Displacement x{i+1}')
    plt.ylabel(f'x{i+1} (m)')
    plt.legend()
    if i == 3:
        plt.xlabel('Time (s)')

plt.suptitle('Response of 4-DOF System to Sinusoidal Force')
plt.tight_layout()
plt.show()
```
